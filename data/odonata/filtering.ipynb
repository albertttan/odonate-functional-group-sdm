```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("odonata.csv", sep='\t')
df = df[df["year"] >= 2020]
df
```

```python
genera = ["Argia", "Aeshna", "Calopteryx", "Hetaerina", "Libellula", "Sympetrum", "Plathemis"]
(min_lat, max_lat, min_lon, max_lon) = (5, 85, -180, -50)
occurences = {}
selected_species = []
```

```python
selected_data = df[["species", "genus", "decimalLatitude", "decimalLongitude"]][(df["decimalLatitude"] > min_lat) & (df["decimalLatitude"] < max_lat) & ((df["decimalLongitude"] > min_lon) & (df["decimalLongitude"] < max_lon))]
selected_data
```

```python
for genus in genera: 
    occurences[genus] = selected_data[selected_data["genus"] == genus].groupby("species").count()["genus"].sort_values(ascending=False)
    selected_species += occurences[genus][occurences[genus] >= 30].iloc[:5].index.to_list()

occurences
```

```python
# selected_species = ["Libellula luctuosa", "Sympetrum corruptum", "Plathemis subornata"]
selected_species
```

```python
with open("selected_species.json", 'w') as f:
    json.dump(selected_species, f)
```
